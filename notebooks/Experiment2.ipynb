```python
import ppmodel
from ppmodel import ModelParams
import numpy as np
from multiprocessing import Pool

n_g = range(5, 51);
n_i = range(3, 21)

args = [[],[]]
result = []
params = []

for g in n_g:
    for i in n_i:
        args[0].append(g)
        args[1].append(i)
        for s in range(0, 3):
            params.append(ModelParams("S", 100, 100, 5, 15, g, i, s))
        
if __name__ == '__main__':
    p = Pool(4)
    resuts = p.map(ppmodel.f, params)
```

```python
maxs = []
mins = []
averages = []

for i in range(2, len(resuts)):
    data = []
    data.append(resuts[i-2])
    data.append(resuts[i-1])
    data.append(resuts[i])
    mins.append(min(data))
    maxs.append(max(data))
    averages.append(np.average(data))
```

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline


fig = plt.figure()
fig.set_size_inches(15, 10)
ax = fig.gca(projection='3d')
ax.plot_trisurf(args[0], args[1], averages, alpha=0.2)
ax.set_xlabel('Amount')
ax.set_ylabel('Interval')
ax.set_zlabel('Steps')
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(15, 10)
ax = fig.gca(projection='3d')
ax.plot_trisurf(args[0], args[1], maxs, alpha=0.2)
ax.set_xlabel('Amount')
ax.set_ylabel('Interval')
ax.set_zlabel('Steps')
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(15, 10)
ax = fig.gca(projection='3d')
ax.plot_trisurf(args[0], args[1], mins, alpha=0.2)
ax.set_xlabel('Amount')
ax.set_ylabel('Interval')
ax.set_zlabel('Steps')
plt.show()
```
