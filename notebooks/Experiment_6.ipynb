```python
import ppmodel
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
```

```python
pred, prey, grass, result = ppmodel.start_model('L', 100, 100, 1, 10, 108, 58, 2)

plt.plot(np.arange(0, len(prey)), prey, 'b', label="Preys")
plt.plot(np.arange(0, len(pred)), pred, 'r', label="Predators")
#plt.plot(np.arange(0, len(grass0)), grass0,'g',label="Grass")
plt.ylabel("Number")
plt.xlabel("Step")
plt.ylim(0, 200)
plt.xlim(1000, 4000)
plt.legend(loc=1)
plt.show()
```
