```python
import ppmodel
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
```

```python
pred, prey, grass, result = ppmodel.start_model('L', 100, 100, 1, 1500, 1500, 35, 178)
plt.figure(figsize=(7.5, 6))
plt.plot(np.arange(0, len(prey)), prey, 'b', label="Herbivore")
plt.plot(np.arange(0, len(pred)), pred, 'r', label="Carnivore")
plt.plot(np.arange(0, len(grass)), grass,'g',label="Plants")
plt.ylabel("Number", fontsize=18)
plt.xlabel("Step", fontsize=18)
#plt.ylim(0, 250)
#plt.xlim(0, 500)
plt.tick_params(labelsize=16)
plt.legend(loc=1, fontsize=16, framealpha=0.5)
plt.savefig("fig61.png", dpi=300)
plt.show()
```
