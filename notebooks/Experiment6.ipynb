```python
from multiprocessing import Pool

import ppmodel
from ppmodel import ModelParams


if __name__ == '__main__':
    p = Pool(4)
    params = []
    for i in range(0, 100):
        params.append(ModelParams("S", 100, 100, 400, 2700, 75, 14, i))
    steps = p.map(ppmodel.f, params)
```

```python
import numpy as np
import scipy.stats as st

steps.sort()
n_bins = int(np.ceil(np.sqrt(len(steps))))
sigma = np.std(steps)
mu = np.mean(steps)
skew = st.skew(steps)
```

```python
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
%matplotlib inline


#plt.figure(figsize=(10, 10))
n, bins, patches = plt.hist(steps, 40,  color="red", alpha=0.7,)

norm_pdf = mlab.normpdf(bins, mu, sigma)

plt.text(2000,70,"Maximum: %i\nMinimum %i\nMean: %f\nSTD: %f\nSkew: %f" % (max(steps), min(steps), mu, sigma, skew),
         size=13, rotation=0,
         ha="left", va="top")

plt.plot(bins, norm_pdf, "r--", label="Normal distribution")
plt.ylabel("Number of cycles")
plt.xlabel("Steps")
plt.legend()
plt.show()

# print("Maximum: %i" % (max(steps)))
# print("Minimum: %i" % (min(steps)))
# print("STD: %f" % (sigma))
# print("Mean: %f" % (mu))
# print("Skew: %f" % (skew))
```
