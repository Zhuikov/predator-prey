# Первый эксперимент

```python
import ppmodel
from ppmodel import ModelParams
import numpy as np
from multiprocessing import Pool

n_r = [1, 5, 15, 20]
n_d = [1, 5, 15, 20]

args = [[],[]]
mins = []
maxs = []
averages = []

for r in n_r:
    for d in n_d:
        args[0].append(r)
        args[1].append(d)
        data = []
        if __name__ == '__main__':
            p = Pool(6)
            params = []
            for s in range(0, 100):
                params.append(ModelParams("S", 100, 100, d, r, 10, 5, s))
            data = p.map(ppmodel.f, params)
        mins.append(min(data))
        maxs.append(max(data))
        averages.append(np.average(data))
```

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline


fig = plt.figure()
fig.set_size_inches(15, 10)
ax = fig.gca(projection='3d')
ax.plot_trisurf(args[0], args[1], averages, alpha=0.2)
ax.set_xlabel('Preys')
ax.set_ylabel('Predators')
ax.set_zlabel('Steps')
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(15, 10)
ax = fig.gca(projection='3d')
ax.plot_trisurf(args[0], args[1], maxs, alpha=0.2)
ax.set_xlabel('Preys')
ax.set_ylabel('Predators')
ax.set_zlabel('Steps')
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(15, 10)
ax = fig.gca(projection='3d')
ax.plot_trisurf(args[0], args[1], mins, alpha=0.2)
ax.set_xlabel('Preys')
ax.set_ylabel('Predators')
ax.set_zlabel('Steps')
plt.show()
```
