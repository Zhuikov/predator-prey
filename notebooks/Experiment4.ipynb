```python
from ppmodel import ModelParams

params = []
args = [[],[],[],[]]
for d in agents2:
    for y in agents1:
            for i in intervals:
                for a in amount:
                        args[0].append(d)
                        args[1].append(y)
                        args[2].append(a)
                        args[3].append(i)
                        params.append(ModelParams("S", 100, 100, d, y, a, i, 2))
print(len(params))
```

```python
import ppmodel

from multiprocessing import Pool 


data = []

if __name__ == '__main__':

    p = Pool(5)

    data = p.map(ppmodel.f, params)
```
