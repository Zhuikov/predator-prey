```python
import numpy as np

amount = 100
interval = 42

preys = np.arange(0 ,2001, 50)
preys[0] = 1
print(preys)

predators = np.arange(0, 251, 10)
predators[0] = 1
print(predators)

seeds = np.array([2, 142, 89])
print(seeds)
```

```python
from ppmodel import ModelParams

params = []
args = [[],[],[]]

for a in preys:
   for i in predators:
        for s in seeds:
            args[0].append(a)
            args[1].append(i)
            args[2].append(s)
            params.append(ModelParams("S", 100, 100, i, a, 100, 42, s))        
print(len(params))
```

```python
import ppmodel
from multiprocessing import Pool 

data = []

try:
    
    file = open("results_e4_1.csv", 'r')
    
except IOError as e:
    print("Launch model")
    if __name__ == '__main__':
        p = Pool(6)
        data = p.map(ppmodel.f, params)

    f = open("results_e4_1.csv", 'w')
    for i in range(0, len(data)):
        f.write(str(args[0][i])+','+str(args[1][i])+','+str(args[2][i])+','+str(data[i])+"\n")
    f.close()
else:
    print("Read file")
    args = [[],[], []]
    for line in file:
        parts = line.split(',')
        args[0].append(int(parts[0]))
        args[1].append(int(parts[1]))
        args[2].append(int(parts[2]))
        data.append(int(parts[3]))
        
    file.close()
```

```python
pltdata = [[],[],[]]
tmpdata = []

for i in range(0,len(data)):
    tmpdata.append(data[i])
    if i%3 == 2:
        pltdata[0].append(args[0][i])
        pltdata[1].append(args[1][i])
        pltdata[2].append(np.average(tmpdata))
        tmpdata = []
        
```

```python
from matplotlib import pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

fig = plt.figure(figsize=(7.5, 6))
ax = fig.gca(projection='3d')
ax.plot_trisurf(pltdata[0], pltdata[1], pltdata[2], alpha=0.15, cmap=cm.Greys)
ax.set_xlabel('Herbivore', fontsize=18, labelpad=15)
ax.set_ylabel('Carnivore', fontsize=18, labelpad=15)
ax.set_zlabel('Steps', fontsize=18, labelpad=20)
ax.tick_params(labelsize=16)
ax.tick_params('z', pad=10)
ax.invert_yaxis()
ax.invert_xaxis()
plt.savefig("fig41.png")
plt.show()
```
