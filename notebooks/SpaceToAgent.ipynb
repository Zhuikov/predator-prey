```python
import ppmodel
from multiprocessing import Pool
from ppmodel import ModelParams

data = [[], []]
number = 100
height = 100
width = 100
square = width * height


for i in range(1, 1 + number):
    data[0].append(i/number) 
    results = []
    if __name__ == '__main__':
        p = Pool(2)
        params = []
        for n in range(0, 10):
            params.append(ModelParams("S", height, width, (square / 2) / number * i, (square / 2) / number * i, n))
        results = p.map(ppmodel.f, params)
        data[1].append(sum(results)/5)
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

ratio = 0
maximum = 0
for i in range(0, len(data[1])):
    if data[1][i] > maximum:
        maximum = data[1][i]
        ratio = data[0][i] 
plt.figure().set_size_inches(10, 10)
plt.plot(data[0], data[1], "bo", alpha=0.6, label="Lifetime")
plt.ylim([0, maximum * 1.1])
plt.xlim([0, max(data[0]) * 1.1])
plt.grid(True)
plt.xlabel("Amount of agents / Square")
plt.ylabel("Steps")
plt.show()

print("Maximum: %f steps, %f ratio" % (maximum, ratio))
```
